```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
```

```python
print("Dataset Shape:", df.shape)
print("\nColumns:")
print(df.columns)

df.info()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop('Cabin', axis=1, inplace=True)

print(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()
print("Duplicates:", duplicates)

df.drop_duplicates(inplace=True)
```

```python
df['Survived'].value_counts().plot(kind='bar')
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
df['Pclass'].value_counts().plot(kind='bar')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(df['Age'], bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Key Findings

1.Missing values in Age and Embarked were handled.
2.Cabin column was removed due to many missing values.
3.Most passengers did not survive.
4.Male passengers formed the majority.
5.Most passengers belonged to Class 3.
6.Passenger ages were concentrated between 20 and 40 years.
